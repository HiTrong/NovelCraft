```python
from diffusers import DiffusionPipeline

pipeline = DiffusionPipeline.from_pretrained("Meina/MeinaPastel_V6")
```

```python
pipeline = pipeline.to("cuda")
```

```python
from flask import Flask, request, send_file
from flask_ngrok import run_with_ngrok
from PIL import Image, ImageDraw, ImageFont
from pyngrok import ngrok
import io

PORT = 5000

app = Flask(__name__)
ngrok.set_auth_token("YOUR-NGROK-AUTH-TOKEN")
public_url = ngrok.connect(PORT).public_url

@app.route('/generate_image', methods=['POST'])
def generate_image():
    data = request.json
    prompt = data.get('prompt', 'Random')
    image = pipeline(prompt).images[0]
    
    img_byte_arr = io.BytesIO()
    image.save(img_byte_arr, format='PNG')
    img_byte_arr.seek(0)
    
    return send_file(img_byte_arr, mimetype='image/png', as_attachment=True, download_name='generated_image.png')

print(public_url+ "/generate_image")
app.run(port=PORT)
```
